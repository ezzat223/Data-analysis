The dataset is too large (around 7.7 million) so i'm reading only 100,000 but the data is divided into years so if we used nrows the data will be biased to only first few years (maybe 1 or 2 only) so i'm picking random rows in an attempt to include large portion of years.

```python
# Set the random seed for reproducibility
random.seed(42)

# Count the number of rows in the CSV file
with open(r"Data/US_Accidents_March23.csv") as f:
    num_rows = sum(1 for line in f)

# Generate a random list of row numbers to skip
# It is sorted as skiprows argument expects to receive a sorted list
skiprows = sorted(random.sample(range(1, num_rows), num_rows - 100000))

# Read a random sample of 100,000 records from the CSV file
accidents = pd.read_csv(r"Data/US_Accidents_March23.csv", skiprows=skiprows)

accidents.to_csv("Data/My random data.csv", index=False)

# Set the display options to show all columns and prevent truncation
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)

accidents.head()
```
